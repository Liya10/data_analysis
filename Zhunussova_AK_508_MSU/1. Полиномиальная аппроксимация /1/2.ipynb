```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
def MyModel (n=1,N=1000,k=100, error=False, mu=0,sigma=0.1, a=10, fun=np.sin):
    
    X=np.linspace(-a*np.pi, a*np.pi,N)
    Y=fun(X)
    
    X_train=np.linspace(-a*np.pi,a*np.pi,k)
    Y_train=fun(X_train)
    if(error == True):
        Y_train+=np.random.normal(mu,sigma,k)
    
    model = np.poly1d(np.polyfit(X_train, Y_train, n))
    
    plt.plot(X,Y,'darkgreen',label="истинное значение")
    plt.plot(X_train,Y_train,'bv',label="эмпирические данные")
    plt.plot(X,model(X),'r',label="обученная модель")
    plt.legend(bbox_to_anchor=(1.1, 1))
    plt.show()
```

```python
def gen_data_9(length=5000):
    x = 2.0 * (np.random.rand(length) - 0.5) * np.pi * 2.0
    f = lambda x: np.exp(-x**2) * (-x) * 5 + x / 3
    y = f(x) + np.random.randn(len(x)) * 0.5
    return x, y, f

fig,ax = plt.subplots(3,3,figsize = (16,16))

x,y,f=gen_data_9(250)

xp = np.linspace(-1, 1, 10000) * 2 * np.pi

for n in range(3):
    for k in range(3):
    
        order = 20*n+8*k
        z = np.polyfit(x,y,order)
        p = np.poly1d(z)

        ax[n,k].scatter(x,y,label = "Real data",s=1)
        ax[n,k].scatter(xp,p(xp),label = "Polynomial with order={}".format(order),
                    color='C1',s=1)
        ax[n,k].legend()
    fig.show()
```
