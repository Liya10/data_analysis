# Ядерное сглаживание для оценки плотности

## Условные обозначения

> N - количество объектов в выборке.

> D — размер признакового пространства, D≥1.

> C — количество классов.

> X={(xi,yi)}Ni=1 — выборка, xi∈RD, yi∈{1,…,C}. x∈RD.

> $X=\{(x_i,y_i)\}_{i=1}^N$ — выборка, где $ x_i\in\mathbb{R}^D,\; y_i\in\{1,..., C\}.$

## Ядерное сглаживание

> Идея: каждый $x_i$ выборки будет центром блока.

> Блок может иметь следующий вид:  $\frac{1}{2} \mathbf{1} \big[\big|\frac{x-x_i}{h}\big|\leq 1\big],\,D=1$

## Одномерный случай $D=1$

> Kernel Density Estimation  (KDE, локальная непараметрическая оценка Парзена-Розенблатта) — 
$\hat{p}(x)=\frac{1}{N\cdot h} \sum_{i=1}^N K\big(\frac{x-x_i}{h}\big)$, $K(\cdot)$ - ядро, чётная и нормированная функция:$\int_{-\infty}^{+\infty}K(u)du=1$.

>Следствие: $\hat{p}(x)$ обладает той же степенью гладкости, что и ядро $K(\cdot)$.

## Виды ядер

- Прямоугольное ядро (tophat kernel):   $\qquad K(u)=\frac {1}{2} \cdot \mathbf{1} [|u|\leq 1]$


- Гауссово ядро:  $\qquad \qquad \qquad \qquad \quad \; K(u)= \frac{1}{\sqrt{2\cdot\pi}} \exp (\frac{-u^2}{2})$


- Ядро Епанечникова:  $\qquad \qquad \qquad \quad \;\; K(u)=\frac{3}{4} \max \{1-u^2,0\}$


- Треугольное ядро: $\qquad \qquad \qquad \qquad K(u)=\max \{1-|u|,0\}$


- Косинусное ядро: $\qquad \qquad \qquad \qquad K(u)=\frac{\pi}{4} \cos(\frac{\pi\cdot u}{2}) \mathbf{1}[|u|\leq 1]$


- Экспоненциальное ядро:  $\qquad\qquad  \quad \; K(u)=\frac{1}{2} \exp (-|u|)$


- Квартическое ядро:   $\qquad \qquad \qquad \quad K(u)=\frac{15}{16} (1-u^2)^2 \mathbf{1}[|u|\leq 1]$

## Эмпирические функции распределения

#### Эмпирическая функция распределения, построенная по выборке $X=(X_1,...,X_n)$ объема n, называется случайная функция $F_n^*:\mathbb{R}\times\Omega\rightarrow[0,1]$, при каждом $y\in \mathbb{R}$ равная 
$$F_n(y)=\frac{\text{количество } X_i\in\{-\infty, y\}}{n}=\sum_{i=1}^n \mathbf{X_i<y}$$

### Теория
#### Эмпирическая функция распределения $F_n(x)$ является эффективной оценкой теорической функции распределения $F(x)$ при каждом x.

## Нормальное распределение
#### формируем данные

```python
n=1000
Contin = st.norm(loc=3, scale=4)
Z_1 = Contin.rvs(size=n)
```

#### сортируем

```python
z_sort=np.sort(Z_1)
```

```python
z_min=z_sort[0]
z_max=z_sort[-1]
```

#### определяем сетку

```python
grid=np.linspace(z_min,z_max,n)
```

### Эмпирическая функция


```python
def empirical(N, element_min, element_max, array_sort):
    grid=np.linspace(element_min,element_max,N)
    cdf=np.zeros(N)
    j=0
    count=1
    n=len(array_sort)
    for i in range(n):
        if(j<N-1):
            if(array_sort[i]<grid[j+1]):
                cdf[j]+=count
                count=1
            else:
                i-=1
                j+=1
                count+=1
        else:
            cdf[j]+=count
            count=1
    if(count>=1):
        cdf[N-1]+=count-1
    cdf=np.cumsum(cdf)
    return [cdf/n, grid]
```

#### ошибка 

```python
def estimate(fun, Grid, CDF):
    n=len(Grid)
    s=0
    for i in range(n):
        s+=(fun(Grid[i])-CDF[i])**2
    return s
```

```python
cdf1, grid1=empirical(N1,z_min,z_max,z_sort)
cdf2, grid2=empirical(N2,z_min,z_max,z_sort)
cdf3, grid3=empirical(N3,z_min,z_max,z_sort)
cdf4, grid4=empirical(N4,z_min,z_max,z_sort)
```

```python
est1=estimate(Contin.cdf, grid1,cdf1)
est2=estimate(Contin.cdf, grid2,cdf2)
est3=estimate(Contin.cdf, grid3,cdf3)
est4=estimate(Contin.cdf, grid4,cdf4)
print("est1=",est1)
print("est2=",est2)
print("est3=",est3)
print("est4=",est4)
```
